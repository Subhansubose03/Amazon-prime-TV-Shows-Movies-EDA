## ***2.Problem Solvings***

### A. Content Distribution by Type

In this section, we analyze the distribution of content between Movies and TV Shows available on Amazon Prime Video. This helps us understand whether the platform is more focused on standalone films or episodic television content. Both count and percentage distribution are examined to provide a clearer comparison.

```python
titles['type'].value_counts()
```

```python
titles['type'].value_counts(normalize=True) * 100
```

####Insight

The analysis shows that Movies dominate Amazon Prime Video’s US content catalog, accounting for approximately 86% of the total titles, while TV Shows represent only about 14%.

This indicates that the platform is heavily skewed toward standalone film content rather than episodic television programming. From a content strategy perspective, this suggests a stronger emphasis on movie acquisition or distribution within the US market.

###B. Genre Dominance Analysis

In this section, we analyze the genre composition of Amazon Prime Video’s US catalog to identify which genres dominate the platform. Understanding genre distribution helps evaluate content diversity and provides insight into audience targeting and strategic content investment.

The objective is to determine:

Which genres appear most frequently in the catalog.

Whether certain genres significantly dominate the platform.

```python
titles['genres'].head()
```

####Data Preparation – Genre Formatting

The genres column is currently stored as string representations of lists.
To accurately analyze genre distribution, we convert these strings into actual Python list objects.

```python
import ast

# Convert string to list
titles['genres'] = titles['genres'].apply(ast.literal_eval)
```

####Expanding Genre Data

Since each title may belong to multiple genres, we expand the dataset so that each genre is treated as an individual entry. This allows accurate frequency calculation.

```python
# Create exploded dataframe
genres_exploded = titles.explode('genres')

# Count genre frequency
genre_counts = genres_exploded['genres'].value_counts()

genre_counts.head(10)
```

####Insight

The analysis reveals that Drama is the most dominant genre in Amazon Prime Video’s US catalog, significantly outperforming all other categories. Comedy and Thriller follow as the second and third most prevalent genres.



###C. Production Country Distribution

In this section, we analyze the distribution of content based on production countries to understand regional contribution and geographic diversity within Amazon Prime Video’s catalog.

```python
titles['production_countries'].head()
```

####Convert to Proper List Format

Even though it already looks like a list, it's still a string representation.

So we convert it properly (just like genres).

```python
# Convert to actual list
titles['production_countries'] = titles['production_countries'].apply(ast.literal_eval)
```

####Explode Country Data

```python
# Explode countries
countries_exploded = titles.explode('production_countries')

# Count frequency
country_counts = countries_exploded['production_countries'].value_counts()

country_counts.head(10)
```

####Insight

The analysis shows that the United States overwhelmingly dominates Amazon Prime Video’s US catalog, contributing more than 5,000 titles. This is expected given the dataset focuses on the US market.

India and the United Kingdom emerge as the next largest contributors, indicating significant international content presence. Countries such as Canada, France, Japan, and Australia also contribute meaningful volumes, suggesting that while the platform is US-heavy, it maintains a globally diverse content portfolio.

###D. Trends Over Time
In this section, we analyze the number of titles released per year to understand how Amazon Prime Video’s content catalog has evolved over time and whether the platform shows signs of expansion.

```python
year_counts = titles['release_year'].value_counts().sort_index()

year_counts.tail(15)
```

####Insight

The data shows a consistent upward trend in content additions from 2008 through 2019, with particularly strong acceleration after 2015. This indicates a significant expansion phase during the mid-to-late 2010s, likely reflecting increasing competition in the streaming industry.

Content growth peaked around 2021, suggesting aggressive content acquisition or production strategy during that period. The lower count observed in 2022 is likely due to incomplete data collection for that year rather than an actual decline.

###E. IMDb Ratings & Popularity

We now answer:

What are the highest-rated shows?

Does higher IMDb score mean higher TMDb popularity?

In this section, we analyze IMDb scores and TMDb popularity metrics to identify the highest-rated titles and examine the relationship between ratings and popularity.

```python
titles[['imdb_score', 'tmdb_score', 'tmdb_popularity']].isnull().sum()
```

####Create Ratings Dataset

To ensure accurate analysis, we filter the dataset to include only titles that have valid IMDb and TMDb scores. This prevents distortion caused by missing rating values.

```python
ratings_df = titles.dropna(subset=['imdb_score', 'tmdb_score'])

ratings_df.shape
```

####Identify Highest-Rated Titles (IMDb)

In this section, we identify the top-rated titles based on IMDb scores to understand which content performs best in terms of audience ratings.

```python
top_imdb = ratings_df.sort_values(by='imdb_score', ascending=False)

top_imdb[['title', 'release_year', 'imdb_score']].head(10)
```

####Insight

The analysis shows that several titles on Amazon Prime Video hold exceptionally high IMDb ratings (above 9.0). Many of these highly rated titles are relatively recent releases (post-2018), indicating strong audience engagement with newer content.

This suggests that high-performing content is not limited to older, classic titles but also includes recent productions, reflecting evolving audience preferences and quality content investments.

####Relationship Between IMDb Score and TMDb Popularity

In this section, we examine whether higher IMDb ratings are associated with higher TMDb popularity scores to understand the relationship between content quality and audience engagement.

```python
correlation = ratings_df['imdb_score'].corr(ratings_df['tmdb_popularity'])
correlation
```

####Insight

The correlation between IMDb score and TMDb popularity is approximately 0.08, indicating a very weak relationship.

This suggests that highly rated content does not necessarily correspond to highly popular content. Audience ratings (quality perception) and popularity metrics (visibility or engagement) appear to function independently.

From a business perspective, this highlights the distinction between critically appreciated content and widely consumed content. High-quality productions may not always achieve maximum popularity, and vice versa

#### Most Popular Shows

```python
top_popular = titles.dropna(subset=['tmdb_popularity']).sort_values(by='tmdb_popularity', ascending=False)

top_popular[['title', 'release_year', 'tmdb_popularity']].head(10)
```

#### Top TMDb Popularity Titles:


All the Old Knives (2022) → 1437.9

Harina (2022) → 951.9

The Eighth Clause (2022) → 950.9

Hotel Transylvania: Transformania (2022) → 934.5

Sonic the Hedgehog (2020) → 893.6

#### Insights

The most popular titles based on TMDb popularity scores are predominantly recent releases, particularly from 2020 to 2022. This indicates that newer content tends to generate higher engagement and visibility.

Unlike IMDb ratings, which reflect audience evaluation of quality, TMDb popularity appears to be strongly influenced by recency and active audience interest. Overall, popularity metrics emphasize current engagement, whereas rating metrics reflect long-term quality perception.

## ***3. Data Visualizations***

### **A. Distribution of Titles by Content Type**

This visualization illustrates the distribution between Movies and TV Shows available on Amazon Prime Video, helping us understand content dominance on the platform.

```python
titles['type'].value_counts().plot(kind='bar')
plt.title("Distribution of Titles by Content Type")
plt.xlabel("Content Type")
plt.ylabel("Number of Titles")
plt.xticks(rotation=0)
plt.show()
```

###  **B. Top 10 Dominant Genres**

This visualization highlights the most prevalent genres in the Amazon Prime Video catalog.

```python
my_colors = ['red', 'green', 'blue', 'cyan', 'magenta', 'yellow', 'lightgreen', 'orange', 'purple', 'brown']

genre_counts.head(10).sort_values().plot(kind='barh', color=my_colors)
plt.title("Top 10 Genres by Frequency")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.show()
```

### **C. Title Count Per Release Year**

This visualization displays the number of titles released per year, illustrating how Amazon Prime’s content library has evolved.

```python
plt.figure(figsize=(10,6))

year_counts.plot(kind='line', color="green")

plt.title("Number of Titles Released per Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.grid(True)

plt.show()
```

### **D. IMDb Scores VS TMDb Popularity**

```python
plt.scatter(ratings_df['imdb_score'], ratings_df['tmdb_popularity'],  c='purple', alpha=0.5)
plt.title("IMDb Score vs TMDb Popularity")
plt.xlabel("IMDb Score")
plt.ylabel("TMDb Popularity")
plt.show()
```

### **E. Actor and Director Participation by Volume**

This section analyzes the participation frequency of actors and directors to understand how talent contribution is distributed across the platform.

```python
# Separate actors and directors
actors = credits[credits['role'] == 'ACTOR']
directors = credits[credits['role'] == 'DIRECTOR']

# Count participation frequency
actor_counts = actors['name'].value_counts()
director_counts = directors['name'].value_counts()

# Display top 10
print("Top 10 Actors by Participation:")
display(actor_counts.head(10))

print("\nTop 10 Directors by Participation:")
display(director_counts.head(10))
```

#### **Visualize Top 10 Actors & Directors Together**

```python
plt.figure(figsize=(14,6))

# Plot actors
plt.subplot(1,2,1)
actor_counts.head(10).plot(kind='bar', color="purple")
plt.title("Top 10 Actors by Participation")
plt.xlabel("Actor")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)

# Plot directors
plt.subplot(1,2,2)
director_counts.head(10).plot(kind='bar',color="green")
plt.title("Top 10 Directors by Participation")
plt.xlabel("Director")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
